## 2.逐块读取文本文件

**对于一个数据量比较大的文件，全部读取是很不方便的。我们通过nrows=参数调整输出的数据的多少**

```python
result = pd.read_csv('PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\ex6.csv')
result
```

```python
#只读取前10行
result = pd.read_csv('PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\ex6.csv',nrows=10)
result
```

```python
#逐块读取文件
chunker =  pd.read_csv('PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\ex6.csv',chunksize=1000)
```

```python
chunker
```

```python
for i in chunker:
    print(i)
```

## 3.将数据写出到文本格式

**数据以csv打开后，再转出csv格式后，数据以逗号分开，当然你可以使用其他的分隔符(使用sep=参数)**

```python
#只读取前10行
data = pd.read_csv('PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\ex5.csv')
data
```

```python
data.to_csv('PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\out5.csv')
```

```python
!type PythonForDataAnalysis-master\PythonForDataAnalysis-master\ch06\out5.csv
```

**ps:每次转存之后，还要使用！type打印出来(Linux使用cat)。我们使用sys在转存出后直接打印出来**

```python
import sys
```

```python
data.to_csv(sys.stdout,sep='|')
```

**有时候数据中有空的元素，我们转存的结果在无元素的地方就空着，为了在空元素的地方添加标识，使用na_rep=**

```python
data.to_csv(sys.stdout,sep='|',na_rep='空的')  #一般空元素使用'Null'
```

**在转存输出的情况下，我们不需要表头，我们可以使用index和header来取消**

```python
data.to_csv(sys.stdout,index=False,header=False)
```

**再来写一个Series的转存机制，里面涉及到了一些关于日期方法**

```python
dates = pd.date_range('1/1/2000',periods=7)
print(dates)
```

```python
ts = pd.Series(np.arange(7),index=dates)
ts.to_csv(sys.stdout)
```
